```python
import cv2
from PIL import Image
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('validated.csv')

df = df[df.valid == True]

df 
```

```python
df = df.drop_duplicates(subset=['mask', 'x', 'y'])

df
```

```python
df['image_exists'] = df['image'].apply(lambda x: os.path.exists(x))
df['mask_exists'] = df['mask'].apply(lambda x: os.path.exists(x))

# total number of images
print('total images:', len(df))
print('total that exist:', df['image_exists'].sum())

# total number of masks
print('total masks:', len(df))
print('total that exist:', df['mask_exists'].sum())
```
