```python
import cv2
from PIL import Image
import os
import numpy as np
import pandas as pd

images_path = '/home/jack/Mounts/DiskOne/kona_coffee/splits'

old_df = None
if os.path.exists('filtered.csv'):
    old_df = pd.read_csv('filtered.csv')

images = [os.path.join(images_path, i) for i in os.listdir(images_path)]

# filter out images that have already been processed
if old_df is not None:
    images = [i for i in images if i not in old_df['image'].values]

print(len(images))

np.random.shuffle(images)
image_index = 0
image = cv2.imread(images[image_index])
data = []
```

```python
cv2.namedWindow('image')

while True:
    
    if image is not None:
        cv2.imshow("image", image)

    k = cv2.waitKey(20) & 0xFF
    
    # if escape key, exit
    if k == 27 or image_index >= len(images):
        break
    
    # if n key, go to next image
    elif k == ord('n'):
        data.append([images[image_index], False])    
        image_index += 1
        if image_index >= len(images):
            break
    
        image = cv2.imread(images[image_index])

    elif k == ord('y'):
        data.append([images[image_index], True])
        image_index += 1
        if image_index >= len(images):
            break
        
        image = cv2.imread(images[image_index])
        
    while image is None:
        print("Image is None")
        data.append([images[image_index], False])
        image_index += 1
        if image_index >= len(images):
            break
        
        image = cv2.imread(images[image_index])


# close all open windows
cv2.destroyAllWindows()
```

```python

df = pd.DataFrame(data, columns=['image', 'keep'])

if old_df is not None:
    df = pd.concat([old_df, df])
    
df = df.drop_duplicates(subset=['image'])

df.to_csv('filtered.csv', index=False)

df
```
